```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.drop('label',axis=1)
y=df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.20, random_state=42)
```

```python
X_train = X_train.values.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.values.reshape(-1, 28, 28, 1) / 255.0
```

```python
y_train_cat=to_categorical(y_train,10)
y_test_cat=to_categorical(y_test,10
                          )
```

```python
perceptron=Sequential((
    Flatten(input_shape=(28,28)),
    Dense(128,activation='relu'),
    Dense(64,activation='relu'),
    Dense(10,activation='softmax')
))
```

```python
perceptron.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
perceptron.fit(X_train,y_train_cat,epochs=5,validation_data=(X_test,y_test_cat))
```

```python
perceptron.evaluate(X_test,y_test_cat,verbose=0)[1]
```
